## İşlemlere geçmeden önce elimizdeki veri setini tanıyalım :

```python
dataf.shape
```

### Bu çıktı bize 1035 gözlem ve 6 değişkenden oluştuğunu söyledi.

### Elimizdeki veri setinin tamamına bir işlem uygulayıp ortalamasını bulalım.

```python
dataf.mean()
```

### Sadece veri setindeki belirli bir değişken için ortalama işlemini uygulamak istersek :

```python
dataf["mass"].mean()
```

### Sadece 'mass' değişkeni için ortalama değeri gösterdi.

### Sadece "mass" değişkenindeki değerleri saymak istersek :

```python
dataf["mass"].count()
```

### Sadece "mass" değişkeni içindekileri karşılaştırarak en küçüğü bulmak istersek :

```python
dataf["mass"].min()
```

### Sadece "mass" içindekileri karşılaştırarak en büyüğü bulmak istersek :

```python
dataf["mass"].max()
```

### Sadece "mass" değişkeni içindekileri değerleri toplamak istersek :

```python
dataf["mass"].sum()
```

### Sadece "mass" değişkeni içindeki değerlerin standart sapmasını bulmak istersek :

```python
dataf["mass"].std()
```

### Sadece "mass" değişkeni içindeki değerlerin varyansını bulmak istersek :

```python
dataf["mass"].var()
```

### Tüm istatistikleri el ile girmemiz gerekmiyor. Bunun için :

```python
dataf.describe()
```

### Bunun tersini yani 'transpoze' halini almak için :

```python
dataf.describe().T
```

### Şu an daha okunaklı bir tablo veri tipi oluştu.

```python
dataf.dropna().describe().T
```

### Veri setindeki eksik verileri sildi. .drop() zaten silme işlemi için kullanılan fonksiyondu. .drop NAN anlamına gelen dropna() fonksiyonu da isminden anlaşıldığı gibi boş veriler sildi.
